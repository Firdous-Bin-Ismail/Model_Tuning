```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
a = df.drop('species', axis=1)
```

```python
b= df['species']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

a = scaler.fit_transform(a)
```

```python
a.astype(int)
```

```python
b
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(a, b, test_size=0.20, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
k_n = KNeighborsClassifier(n_neighbors = 5 )
```

```python
k_n.fit(X_train, y_train)
```

```python
k_n.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC(gamma='auto')
```

```python
model_svc.fit(X_train, y_train)
```

```python
model_svc.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV

```

```python
A = GridSearchCV((model_svc),{
    'C' : [5,10,25,30],
    'kernel' : ['linear', 'rbf'],

    }, cv =5, return_train_score=False)
```

```python
A.fit(a,b)
```

```python
A.cv_results_
```

```python
results = pd.DataFrame(A.cv_results_)
```

```python
results
```

```python
results [['param_C', 'param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
B = RandomizedSearchCV((model_svc),{
    'C' : [5,10,25,30],
    'kernel' : ['linear', 'rbf'],

    }, n_iter=4, cv =5, return_train_score=False)
```

```python
B.fit(a,b)
```

```python
B.cv_results_
```

```python
res = pd.DataFrame(B.cv_results_)
```

```python
res
```

```python
res [['param_C', 'param_kernel','mean_test_score']]
```
